```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('heart_disease_health_indicators_BRFSS2015 2 (1) (1).csv')
df.head()
```

```python
plt.figure(figsize=(10,4))
sns.histplot(df['Age'], bins=13, kde=True, color='orange')
plt.title('Age Distribution', size=20, color='red')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['BMI'], bins=20, kde=True, color='maroon')
plt.title('BMI Distribution', size=20, color='red')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['MentHlth'], bins=20, kde=True)
plt.title('Mental Health Days Distribution', size=20, color='red')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['PhysHlth'], bins=20, kde=True, color='green')
plt.title('Physical Health Days Distribution', color='red', size=20)
plt.show()
```

```python
sns.countplot(x='HighBP', data=df, hue='HighBP')
plt.title('Prevalance of High Blood Pressure', color='red', size=20)
plt.show()
```

```python
sns.countplot(x='HighChol', data=df, hue='HighChol')
plt.title('Prevalence of High Cholesterol', color='red', size=20)
plt.show()
```

```python
Diabetes=df['Diabetes'].value_counts()
plt.figure(figsize=(7,4))
plt.pie(Diabetes, labels=Diabetes.index, autopct='%1.1f%%', colors=['green', 'orange', 'skyblue'], shadow=True)
plt.title('Prevalence of Diabetes', color='red', size=20)
plt.show()
```

```python
Stroke=df['Stroke'].value_counts()
plt.figure(figsize=(8,4))
plt.pie(Stroke, labels=Stroke.index, autopct='%1.1f%%', colors=['yellow','orange'], shadow=True)
plt.title('Prevalence of Stroke', color='red', size=20)
plt.show()
```

```python
Smoker=df['Smoker'].value_counts()
plt.figure(figsize=(9,4))
plt.pie(Smoker, labels=Smoker.index, autopct='%1.1f%%', wedgeprops=dict(width=0.5), colors=['orange','green'], shadow=True)
plt.title('Prevalance of Smoker', color='red', size=20)
plt.show()
```

```python
Hartdiseas=df['HeartDiseaseorAttack'].value_counts()
plt.figure(figsize=(5,5))
plt.pie(Hartdiseas, labels=Hartdiseas.index, autopct='%1.1f%%', colors=['red','green'], shadow=True)
plt.title('Percentage of Heart Disease Cases', color='red', size=20)
plt.show()
```

```python
sns.countplot(x='HighBP', hue='HeartDiseaseorAttack', data=df)
plt.title('High BP Vs Hart Diseas', color='red', size=20)
plt.xlabel('High BP', color='blue', size=15)
plt.ylabel('Hart Diseas', color='blue', size=15)
plt.show()
```

```python
sns.boxplot(x='HeartDiseaseorAttack', y='BMI', data=df, hue='HeartDiseaseorAttack')
plt.title('BMI vs Heart Disease', color='red', size=20)
plt.xlabel('HeartDiseaseorAttack', color='blue', size=15)
plt.ylabel('BMI', color='blue', size=15)
plt.show()
```

```python
sns.countplot(x='HighChol', hue='HeartDiseaseorAttack', data=df)
plt.title('HighChol Vs Hart Disease', color='red', size=20)
plt.xlabel('HighChol', color='blue', size=20)
plt.show()
```

```python
sns.countplot(x='Age', hue='HeartDiseaseorAttack', data=df)
plt.title('Age vs Heart Disease', color='red', size=20)
plt.xlabel('Smoker', color='blue', size=15)
plt.show()
```

```python
sns.countplot(x='Diabetes', hue='HeartDiseaseorAttack', data=df)
plt.title('Diabetes vs Heart Disease', color='red', size=20)
plt.xlabel('Diabetes', color='blue', size=15)
plt.show()
```

```python
physical_activity=df.groupby('PhysActivity')['HeartDiseaseorAttack'].sum()
plt.pie(physical_activity, labels=physical_activity.index, autopct='%1.1f%%', shadow=True)
plt.title('PhysActivity Vs Hart Disease', color='red', size=20)
plt.show()
```

```python
sns.countplot(x='Education', hue='HeartDiseaseorAttack', data=df)
plt.title('Education vs Heart Disease', color='red', size=20)
plt.xlabel('Education', color='blue', size=15)
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```
