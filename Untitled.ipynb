```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('heart_disease_health_indicators_BRFSS2015 2 (1) (2).csv')
df.head()
```

```python
df['PhysActivity'].value_counts().plot.pie(autopct='%1.1f%%', labels=['No Activity', 'Activity'],
                                         figsize=(5,5), shadow=True)
plt.title('Physical Activity Percentage', color='red', size=20)
plt.show()
```

```python
sns.countplot(x='Smoker', data=df, hue='Smoker')
plt.title('Smoking Status Distribution', color='red', size=20)
plt.xlabel('Smoker (0=No, 1=Yes)', color='blue', size=15)
plt.show()
```

```python
df['Fruits'].value_counts().plot.pie(labels=['No Fruits', 'Consume Fruits'], autopct='%1.1f%%',
                                     figsize=(5,5), shadow=True)
plt.title('Fruit Consumption Percentage', color='red', size=20)
plt.show()
```

```python
sns.countplot(x='Veggies', data=df, hue='Veggies')
plt.title('Vagetable Consumption Distribution', color='red', size=20)
plt.xlabel('Veggies (0=No, 1=Yes)', color='blue', size=15)
plt.show()
```

```python
sns.countplot(x='HvyAlcoholConsump', data=df, hue='HvyAlcoholConsump')
plt.title('Alcohole Consumption Distribution', color='red', size=20)
plt.xlabel('Consume Alcohole (0=No, 1=Yes)', color='blue', size=15)
plt.show()
```

```python
import math
cols=['Smoker', 'PhysActivity', 'Fruits', 'Veggies', 'HighBP', 'HighChol', 'Diabetes']
n=len(cols)
rows=math.ceil(n/2)
fig, axes=plt.subplots(rows, 2, figsize=(8, 4*rows))
for ax, col in zip(axes.flatten(), cols):
    stack=pd.crosstab(df[col], df['HeartDiseaseorAttack'])
    stack.plot(kind='bar', stacked=True, ax=ax)
    ax.set_title(f'{col} vs Heart Disease')
    ax.set_xlabel(col)
    ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
sns.histplot(df['BMI'], bins=20, color='green')
plt.title('BMI Distribution', color='red', size=20)
plt.show()
```

```python
sns.histplot(df['MentHlth'], bins=30)
plt.title('Mental Health Days Distribution', color='red', size=20)
plt.show()
```

```python
sns.histplot(df['PhysHlth'], color='orange', bins=10)
plt.title('Physical Health Days Distribution', color='red', size=20)
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap', color='red', size=20)
plt.show()
```

```python
sns.scatterplot(x='BMI', y='PhysHlth', hue='HeartDiseaseorAttack', data=df)
plt.title('BMI vs Physical Health Days', color='red', size=20)
plt.show()
```

```python
age=df.groupby('Age')['HeartDiseaseorAttack'].mean().reset_index()
sns.lineplot(x='Age', y='HeartDiseaseorAttack', marker='o', data=age)
plt.title('Heart Disease Trend Across Age', color='red', size=20)
plt.xlabel('Age', color='blue', size=15)
plt.ylabel('Heart Disease Rate', color='blue', size=15)
plt.show()
```

```python
sns.violinplot(x='HeartDiseaseorAttack', y='BMI', data=df)
plt.title('BMI Distribution by Heart Disease', color='red', size=20)
plt.show()
```

```python
sns.pairplot(df[['MentHlth', 'PhysHlth', 'BMI', 'HeartDiseaseorAttack']], hue='HeartDiseaseorAttack')
plt.show()
```
